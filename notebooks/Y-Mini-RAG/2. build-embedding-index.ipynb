```python
import json

with open("chunks.json") as f:
    data = json.load(f)

all_chunks = [d["text"] for d in data]
```

```python
from embed_model_utils import encode
chunk_embeddings = encode(all_chunks).numpy()

print("chunk_embeddings shape:", chunk_embeddings.shape)
```

```python
import joblib

index_data = {
    "chunks": all_chunks,
    "embeddings": chunk_embeddings,
}

joblib.dump(index_data, "rag_index.pkl")
print("saved: rag_index.pkl")
```
