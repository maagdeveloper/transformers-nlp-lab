```python
from embed_model_utils import encode
import joblib

index_data = joblib.load("rag_index.pkl")

all_chunks = index_data["chunks"]
chunk_embeddings = index_data["embeddings"]

print("chunks:", len(all_chunks))
print("embeddings shape:", chunk_embeddings.shape)
```

```python
import numpy as np
def retrieve(query, top_k=2):
    query_emb = encode([query]).numpy()[0]

    scores = np.dot(chunk_embeddings, query_emb)

    top_indices = np.argsort(scores)[::-1][:top_k]

    results = []
    for idx in top_indices:
        results.append({
            "chunk_id": int(idx),
            "score": float(scores[idx]),
            "text": all_chunks[idx]
        })

    return results
```

```python
queries = [
    "What is RAG?",
    "Explain BERT",
    "What is tokenization?",
    "Define machine learning"
]

for query in queries:
    results = retrieve(query, top_k=2)

    print("query:", query)
    for r in results:
        print("chunk_id:", r["chunk_id"])
        print("score   :", round(r["score"], 4))
        print("text    :", r["text"])
        print("-" * 40)
    print("=" * 60)
```
